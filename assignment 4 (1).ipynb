# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = 
https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views 

```python
import re

# selenium
import selenium
import pandas as pd
from selenium import webdriver

# Beautiful soup
from bs4 import BeautifulSoup
import requests

# add time
import time

from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException

from selenium.webdriver.support.ui import WebDriverWait
```

```python
# first, connect to the webdriver
driver=webdriver.Chrome()

# getting the webpage of mentioned url
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
Rank = []
Name = []
Artist = []
Date = []
Views = []
```

```python
# scraping Rank of the videos
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]"):
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("-")
        
```

```python
# Scraping Name of the videos
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]"):
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-")
        
# Scraping Artist of the videos
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]"):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")
        
# Scraping Upload_Date of the videos
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]"):
        Date.append(i.text)
except NoSuchElementException:
    Date.append("-")
        
# Scraping Views of the videos
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]"):
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")
```

```python
Wiki = pd.DataFrame({})
Wiki['Rank'] = Rank
Wiki['Name'] = Name
Wiki['Artist'] = Artist
Wiki['Upload Date'] = Date
Wiki['Views (in Billions)'] = Views

# removing stray numbers from Name column
Wiki.Name = Wiki.Name.apply(lambda x:x[:-4].strip('"'))
Wiki
```

# 4. Scrape the details of trending repositories on Github.com. 
Url = https://github.com/
You have to find the following details: 
A) Repository title 
B) Repository description 
C) Contributors count 
D) Language used

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

url = 'https://github.com/trending'

# Set up the webdriver
driver = webdriver.Chrome()
driver.get(url)

# Find the repository elements
repositories = driver.find_elements(By.XPATH, '//article[@class="Box-row"]')

# Loop through the repositories and extract the details
for repository in repositories:
    try:
        title = repository.find_element(By.XPATH, './/h1/a').text.strip()
    except NoSuchElementException:
        title = 'N/A'
    
    try:
        description = repository.find_element(By.XPATH, './/p').text.strip()
    except NoSuchElementException:
        description = 'N/A'
    
    contributors = len(repository.find_elements(By.XPATH, './/a[@data-hovercard-type="user"]'))
    
    try:
        language = repository.find_element(By.XPATH, './/span[@itemprop="programmingLanguage"]')
        language = language.text.strip()
    except NoSuchElementException:
        language = 'N/A'
    
    # Print the details
    print(f'Title: {title}')
    print(f'Description: {description}')
    print(f'Contributors: {contributors}')
    print(f'Language: {language}')
    print('---')

# Close the webdriver
driver.quit()

```

# 7. Scrape the details most watched tv series of all time from imdb.com. 
Url = https://www.imdb.com/list/ls095964455/ You 
have to find the following details: 
A) Name 
B) Year span 
C) Genre 
D) Run time 
E) Ratings 
F) Votes

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

url = 'https://www.imdb.com/list/ls095964455/'

driver = webdriver.Chrome()
driver.get(url)

shows = driver.find_elements(By.XPATH, '//div[@class="lister-item mode-detail"]')

for show in shows:
    name = show.find_element(By.XPATH, './/h3[@class="lister-item-header"]/a')
    year_span = show.find_element(By.XPATH, './/span[@class="lister-item-year text-muted unbold"]')
    genre = show.find_element(By.XPATH, './/span[@class="genre"]')
    run_time = show.find_element(By.XPATH, './/span[@class="runtime"]')
    ratings = show.find_element(By.XPATH, './/div[@class="ipl-rating-star small"]//span[@class="ipl-rating-star__rating"]')
    votes = show.find_element(By.XPATH, './/span[@name="nv"]')

    print(f'Name: {name.text}')
    print(f'Year span: {year_span.text}')
    print(f'Genre: {genre.text}')
    print(f'Run time: {run_time.text}')
    print(f'Ratings: {ratings.text}')
    print(f'Votes: {votes.text}')
    print()

driver.quit()

```

# 5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/
You have to find the following details: 
A) Song name 
B) Artist name 
C) Last week rank 
D) Peak rank 
E) Weeks on board 
Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
import requests
from bs4 import BeautifulSoup

def get_billboard_hot_100(year):
    date = f"{year}-01-01"
    url = f"https://www.billboard.com/charts/hot-100/{date}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    song_tags = soup.find_all("h3", class_="c-title a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only")
    songs = []
    for tag in song_tags:
        song_text = tag.get_text().strip()
        songs.append(song_text)
    return songs

year = input("Enter a year (YYYY): ")
print(get_billboard_hot_100(year))
```

# 8)Details of Datasets from UCI machine learning repositories. 
Url = https://archive.ics.uci.edu/
You have to find the following details: 
A) Dataset name 
B) Data type 
C) Task 
D) Attribute type 
E) No of instances 
F) No of attribute G) Year 
Note: - from the home page you have to go to the Show All Dataset page through code.

```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
driver = webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")

show_all_datasets = driver.find_element(By.XPATH,'/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b')
show_all_datasets.click()

table = driver.find_element(By.XPATH,'//*[@id="content"]/table[1]')
rows = table.find_elements(By.XPATH,'.//tr')

for row in rows:
    cells = row.find_elements(By.XPATH,'.//td')
    if len(cells) == 7:
        dataset_name = cells[0].text
        data_type = cells[1].text
        task = cells[2].text
        attribute_type = cells[3].text
        no_of_instances = cells[4].text
        no_of_attributes = cells[5].text
        year = cells[6].text
        print(dataset_name, data_type, task, attribute_type, no_of_instances, no_of_attributes, year)

driver.quit()
```

# 2)Scrape the details team India‚Äôs international fixtures from bcci.tv. 
Url = https://www.bcci.tv/.
You need to find following details: 
A) Match title (I.e. 1 ODI) 
B) Series 
C) Place 
D) Date 
E) Time 
Note: - From bcci.tv home page you have reach to the international fixture page through code. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Set up the driver
driver = webdriver.Chrome()
driver.get("https://www.bcci.tv/international/fixtures")

# Scrape the data
table = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class='js-list']/a")))
table.click()
fixtures = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class='fixture__format-strip']")))
fixtures_html = fixtures.get_attribute('outerHTML')
df = pd.read_html(fixtures_html)[0]

# Clean up the data
df.columns = df.iloc[0]
df.drop(0, inplace=True)
df.reset_index(drop=True, inplace=True)

# Print the data
print(df[['Match title', 'Series', 'Place', 'Date', 'Time']])
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Set up the driver
driver = webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_best-selling_books")

# Scrape the data
table = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//table[@class='wikitable sortable']")))
df = pd.read_html(table.get_attribute('outerHTML'))[0]

# Clean up the data
df.columns = df.iloc[0]
df.drop(0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.rename(columns={'Author(s)': 'Author name', 'Publisher': 'Publisher name'}, inplace=True)

# Print the data
print(df[['Book name', 'Author name', 'Volumes sold', 'Publisher name', 'Genre']])
```

# 3 Scrape the details of State-wise GDP of India from statisticstime.com. 
Url = http://statisticstimes.com/
You have to find following details: A)
Rank 
B) State 
C) GSDP(18-19)- at current prices 
D) GSDP(19-20)- at current prices 
E) Share(18-19) 
F) GDP($ billion) 
Note: - From statisticstimes home page you have to reach to economy page through code.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Set up the driver
driver = webdriver.Chrome()
driver.get("http://statisticstimes.com/")

# Navigate to the Economy page
economy_link = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.LINK_TEXT, "Economy")))
economy_link.click()

india_gdp_link = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.LINK_TEXT, "India")))
india_gdp_link.click()

statewise_gdp_link = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.LINK_TEXT, "State-wise GDP")))
statewise_gdp_link.click()

# Scrape the data
table = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//table[@id='table_id']")))
df = pd.read_html(table.get_attribute('outerHTML'))[0]

# Clean up the data
df.columns = df.iloc[0]
df.drop(0, inplace=True)
df.reset_index(drop=True, inplace=True)

# Print the data
print(df[['Rank', 'State', 'GSDP(18-19)- at current prices', 'GSDP(19-20)- at current prices', 'Share(18-19)', 'GDP($ billion)']])
```
