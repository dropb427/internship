#1) Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
first_headertag=soup.find('span',class_="mw-headline")
```

```python
first_headertag
```

```python
first_headertag.text
```

```python
headertag= []
for i in soup.find_all('span',class_="mw-headline"):
    headertag.append(i.text)
    
headertag
```

```python
import pandas as pd
df=pd.DataFrame({'headertag':headertag})
```
